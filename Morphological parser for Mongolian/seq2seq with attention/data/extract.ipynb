```python
 NOUN_INFLECTION = [
  [ "ийн", 'ын', 'ы', 'ий', 'н' ],
  [ 'д', 'т' ],
  [ 'ыг', 'ийг', 'г' ],
  [ 'аас', 'ээс', 'оос', 'өөс' ],
  [ 'аар', 'ээр', 'оор', 'өөр' ],
  [ 'тай', 'тэй', 'той' ],
  [ 'руу', 'рүү', 'луу', 'лүү' ],                  
 ] 
```

```python
def formatRow(data, sourceWord, list):
  for row in data:
    list.append( {'src': sourceWord, 'der': row[0], 'inf': row[1], 'parse': sourceWord + ' + ' + row[1]} )
```

```python
def getData(row):
  returnList = [];
  inflections = row['Inflection']
  wordList = inflections.split(", ")
  for word in wordList:
    element = getInflection(word)
    print(word, element)
    if element is None: 
      inf = getSubStr(row['lemma'], word, True)
      print(inf)
      element = [word, inf]
    returnList.append( element )
  return returnList

def getInflection(word):
  for inflection in NOUN_INFLECTION:
    for inf in inflection:
      infFromWord = getSubStr(inf, word)
      if infFromWord == inf :
        return [word, inf]

def getSubStr(subWord, word, bool = False):
  subLen = len(subWord)
  wordLen = len(word)
  if bool :
    return word[ subLen : wordLen ]
  return word[ (wordLen - subLen) : wordLen ]
```

```python
import pandas as pd

#  **Get all words from excel (data frame)**
excel = pd.read_csv('words.tsv',sep="\t",header=0)
data = pd.DataFrame(excel, columns = ['word_id', 'lemma', 'POS', 'Category', 'Derivation', 'Source word', 'Derived POS', 'Source POS', 'Inflection'])

list = []
for index, row in data.iterrows():
  if type(row['Inflection']) is str :   
    parsedData = getData(row)
    formatRow(parsedData, row['lemma'], list)

# print(list)
#export
# df = pd.DataFrame(list)
# df.to_csv('mn-noun.tsv', sep = '\t', index=False, encoding='utf-8')
```
