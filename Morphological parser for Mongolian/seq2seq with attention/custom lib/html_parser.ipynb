```python
from bs4 import BeautifulSoup
import requests
```

### Send request and get formatted HTML
@url - URL of request

```python
def getFormattedHtml(url):
  res = getRequests(url)
  return BeautifulSoup(res, 'lxml')
```

### Send request
@url - URL of request

```python
def getRequests(url):
  res = requests.get(url)
  if res != None : return res.text
```

### Find from html by tag name and class name
@tag = tag name
@clasName = class name

```python
def findAll(file, tag, className = False):
 if className is False: return file.findAll(tag)
 return file.findAll(tag, class_ = className)
```

```python
def find(file, tag, className = False):
  if className is False: return file.find(tag)
  return file.find(tag, class_ = className)
```
